# üìè What is the Line Equation?
The line equation in 2D is written as:

ùë¶ = ùëöùë• + ùëê

Where:

y is the output (dependent variable)

x is the input (independent variable)

m is the slope of the line (how steep it is)

c is the y-intercept (where the line crosses the y-axis)

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values
x = np.linspace(-10, 10, 100)

# Different lines with different slopes and intercepts
lines = [
    {'m': 1, 'c': 0, 'label': 'y = 1x + 0'},
    {'m': 2, 'c': 1, 'label': 'y = 2x + 1'},
    {'m': -1, 'c': 2, 'label': 'y = -1x + 2'},
    {'m': 0.5, 'c': -3, 'label': 'y = 0.5x - 3'}
]

plt.figure(figsize=(10, 6))
for line in lines:
    y = line['m'] * x + line['c']
    plt.plot(x, y, label=line['label'])

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Understanding Line Equation: y = mx + c")
plt.xlabel("x (input)")
plt.ylabel("y (output)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
