## Linear Regression is apply when you have single feature to predict

#### 1. **Linear Regression**
- Models the linear relationship between the dependent and independent variable(s).
- Equation:  
  $
  Y = \beta_0 + \beta_1X + \varepsilon
  $

  or

  Y = mx + c (This is what we learn in school in line equation, where m is slop of the line with x axis)
- Example: Predicting house prices based on size.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# import the dataset
dataset = pd.read_csv('cgpa_vs_package_dataset.csv')
dataset.head(3)
```

```python
#check null, if available then fillna
dataset.isnull().sum()
```

```python
# check if our dataset follow the linearity 
plt.figure(figsize=(5,3))
sns.scatterplot(x = "CGPA", y ="Package_LPA", data=dataset)
plt.show()
```

```python
# preparing the input and output variable 
x = dataset[['CGPA']] # x should be 2d data for machine learning
y = dataset['Package_LPA']
```

```python
# now split data in train and test datasets

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=65) # random_state ?
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# linear regression will give the best values for m and c for y= mx+c
lr = LinearRegression()
lr.fit(x_train, y_train) # now model is trained
```

```python
# check the accuracy score
lr.score(x_test, y_test)
```

```python
# check the value of m and c provide by lr
lr.coef_, lr.intercept_
```

```python
# now verify the m and c for 6.87	11.77 where x = 6.87
lr.coef_ * 6.87 + lr.intercept_ # verify this from next predict method
```

```python
# now test the model, using 6.87	11.77
lr.predict([[6.87]])
```

```python
# now draw the line graph in the plot
plt.figure(figsize=(7,3))
sns.scatterplot(x = "CGPA", y ="Package_LPA", data=dataset)
plt.plot(dataset['CGPA'], lr.predict(x), c = 'red')
plt.legend(['org', 'predict line'])
plt.show()
```
