```python
# importing the required packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing the data set
dataset = pd.read_excel('Financial Sample.xlsx')
dataset.head(3)
```

```python
# now find the null values (that is in NaN form in dataset)
dataset.isnull().sum(), dataset.isnull().sum().sum()
```

```python
# graph
sns.heatmap(dataset.isnull())
plt.show()
```

```python
# data deleting
# if there is more then 50% of data is missing in a column then delete that column

dataset.drop(columns=["Discount Band"], inplace=True) # here we deleting the entire column
```

```python
# now deleting the null values data only.
dataset.dropna(inplace=True) #it deletes the Rows with missing values
```
