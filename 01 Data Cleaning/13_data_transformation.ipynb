Transforming/convert the mix data type(object) to the relevant data type

```python
import pandas as pd
```

```python
# loading the data
dataset = pd.read_csv("SuperStoreUS.csv", encoding="latin1")
dataset.head(3)
```

```python
#now getting the information of the dataset
dataset.info()
```

```python
# lets take the Unit Price column for this example

#First check the null value in it, if available then fill with the mean or mode
dataset.isnull().sum()
dataset['Unit Price'].isnull().sum()
```

```python
# now fill na with mean

dataset['Unit Price'].fillna(dataset['Unit Price'].mean(), inplace=True)
dataset.head(3)
```

```python
#now converting the data type
dataset['Unit Price'] = dataset['Unit Price'].astype('float32')
dataset.info()
```
