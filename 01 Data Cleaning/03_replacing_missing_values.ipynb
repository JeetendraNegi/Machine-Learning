```python
# importing the required packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing the data set
dataset = pd.read_excel('Financial Sample.xlsx')
dataset.head(3)
```

```python
# now find the null values (that is in NaN form in dataset)
dataset.isnull().sum(), dataset.isnull().sum().sum()
```

```python
# filling the null values
dataset.fillna(10) #filling 10 in place of null values

# this is not right way.
```

```python
# backfilling the data
dataset.fillna(method='bfill') #it will backfill the data like (if in a column there is null value in row 2 and row 3 have value then it will fill the row 3 value to row 2)
```

```python
# forword filling
dataset.fillna(method='ffill') # same as bfill but i will fill the data from next data
```

```python
dataset.info()
```

```python
# filling mode 
#first collect the data base on data type, 
dataset.select_dtypes(include='object').isnull().sum()
```

```python
dataset.select_dtypes(include='number').isnull().sum()
```

```python
#filling the mode for all the columns with mode

for column in dataset.select_dtypes(include='object').columns:
    mode = dataset[column].mode()[0]
    #filling the mode
    dataset[column].fillna(mode, inplace=True)

for column in dataset.select_dtypes(include='number').columns:
    mode = dataset[column].mode()[0]
    #filling the mode
    dataset[column].fillna(mode, inplace=True)
```
