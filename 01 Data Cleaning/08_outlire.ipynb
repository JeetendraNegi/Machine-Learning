**Definition:**
An **outlier** is a data point that significantly deviates from other observations in a dataset. Outliers can arise due to variability in the data, measurement errors, or experimental anomalies. Identifying outliers is crucial as they can influence statistical analyses and potentially lead to misleading interpretations. :contentReference[oaicite:0]{index=0}

**Identifying Outliers Using the Interquartile Range (IQR):**
A common method to detect outliers is the **1.5 * IQR rule**. The IQR measures the middle 50% of a dataset and is calculated as:

IQR = Q3 - Q1


Where:
- `Q1` = First quartile (25th percentile)
- `Q3` = Third quartile (75th percentile)

Data points are considered outliers if they fall below:
