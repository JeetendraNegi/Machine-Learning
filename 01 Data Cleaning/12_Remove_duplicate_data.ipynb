# This is not mean duplicate data in a column, its mean entire row is duplicate.

```python
import pandas as pd
```

```python
# data preparation 
data = {
    'name':[
        'a', 'b', 'c', 'd', 'a', 'c'
    ],
    'eng':[
        8,7,5,8,8,4
    ],
    'hindi':[
        2,3,4,5,2,6
    ]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# check the duplicates
df.duplicated()
```

```python
# adding this columns in df
df["duplicates"] = df.duplicated()
df
```

```python
# removing the duplicate rows

# first remove the duplicates column
df.drop('duplicates', axis=1, inplace=True)
```

```python

df.drop_duplicates(keep="first")
```

```python
df.drop_duplicates(inplace=True)
df
```

# Working with the Real data

```python
#importing the data
dataset = pd.read_csv("SuperStoreUS.csv", encoding='latin1')
dataset.head(3)
```

```python
# checking the rows and columns
dataset.shape
```

```python
# checking the duplicate data
dataset.duplicated()

# adding a new columns with duplicates
dataset["duplicates"] = dataset.duplicated()
dataset.head(3)
```

```python
# try to getting the only duplicate rows
duplicate_dataset = dataset[dataset.duplicated(keep=False)]
duplicate_dataset
```

```python
# now delete the duplicate
# first remove the duplicate columnn
dataset.drop('duplicates', axis=1, inplace=True)
dataset.head(3)
```

```python
# removing the duplicates
dataset.drop_duplicates(keep="first", inplace=True)
dataset.shape
```
