```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the dataset

dataset = pd.read_csv('age_experience_salary_dataset.csv')
dataset.head(5)
```

```python
# check the null
dataset.isnull().sum()
```

```python
# input and output variables
x = dataset.iloc[:, :-1]
y = dataset['Salary']
```

```python
# train test and split data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2, random_state= 42)
```

```python
# model selection

from sklearn.tree import DecisionTreeRegressor

dtR = DecisionTreeRegressor(criterion="absolute_error", max_depth=4, splitter="random")
```

```python
# train the model
dtR.fit(x_train, y_train)
```

```python
# check the accuracy
dtR.score(x_test, y_test)
```

```python
# check the overfitting 

dtR.score(x_train, y_train)
```

```python
# this is fully overfitting case
```

## HyperParameter Tuning using GridSearchCV and RandomezedSearchCV

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
# creating the list of Hyperparameters
df = {"criterion": ["squared_error", "friedman_mse", "absolute_error", "poisson"],
      "splitter": ["best", "random"],
      "max_depth": [i for i in range (2, 20)]}
```

```python
gridSearchCV = GridSearchCV(DecisionTreeRegressor(), param_grid=df)
gridSearchCV.fit(x_train, y_train)
```

```python
# check the best parameters
gridSearchCV.best_params_
```

```python
gridSearchCV.score(x_test, y_test)
```

```python
# using these values in the DecisionTreeRegression as parameters and check the accuracy
```

```python
randomizedSearchCV = RandomizedSearchCV(DecisionTreeRegressor(), param_distributions=df, n_iter=20)
randomizedSearchCV.fit(x_train, y_train)
```

```python
randomizedSearchCV.best_params_
```

```python
randomizedSearchCV.best_score_
```
