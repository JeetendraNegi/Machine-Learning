## Agglomerative Clustering

## Agglomerative clustering only work on linearly separable data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the dataset
df = pd.read_csv("iris.csv")
df.head(5)
```

```python
# now remove the output from this and make a new dataset

dataset = df.iloc[:, :-1]
dataset.head(3)
```

```python
# check the linearly separable graph

sns.pairplot(data=dataset)
plt.show()
```

## Now create a Dendrogram to figure out how many Cluster will be ok for this

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20, 20))
sc.dendrogram(sc.linkage(dataset, method="single", metric='euclidean'))
plt.show()
```

```python
# now we'll create the 2 cluster 
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2, linkage="single")
ac.fit_predict(dataset)
```

```python
# now create a new column in dataset

dataset['AC_Predict'] = ac.fit_predict(dataset)
```

```python
# now create a pair plot with predict

sns.pairplot(data=dataset, hue='AC_Predict')
plt.show()
```
